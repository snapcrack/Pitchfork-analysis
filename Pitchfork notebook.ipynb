```python
##WORD CLEANING

'''added .lower() in .append() in clean'''

def clean(x):
    cleanwords = []
    for word in x[:]:
        word = update_words(word)
        cleanwords.append(word)
    return cleanwords

def update_words(word):
    clean_word = re.findall(r'(.*)[.]', word)
    if clean_word:
        word = clean_word[0]
    clean_word2 = re.findall(r'[A-Z].*', word)
    if clean_word2:
        word = clean_word2[0]
    clean_word3 = re.findall(r'^\W*(\w*)', word)
    if clean_word3:
        word = clean_word3[0].lower
    clean_word4 = re.findall(r'(\w*)\W*$', word)
    if clean_word4:
        word = clean_word4[0]
    return word.lower()
'''
    if clean_word:
        cleaned_word = clean_word[0]
        return cleaned_word
    if clean_word2:
        cleaned_word2 = clean_word2[0].lower()
        return cleaned_word2
    if clean_word3:
        cleaned_word3 = clean_word3[0].lower()
        return cleaned_word3
    if clean_word4:
        cleaned_word4 = clean_word4[0].lower()
        return cleaned_word4
    else:
        return word.lower()
'''
```

```python
###COUNTING WORDS

def counting(x):
    counter = {}
    for e in x:
        if e not in counter:
            counter[e] = 0
        if e in counter:
            counter[e] += 1
    return counter
```

```python
import operator
```

```python
reviews['counter'] = reviews['cleanwords'].apply(counting)
```

```python
## GETTING PROPORTIONS OF WORD COUNTS

def props(x):
    proportions = {}
    y = 0
    while y < len(reviews):
        for key, val in x.items():
            val = float(val)
            prop = float(val / len(reviews['content'][x]))
            proportions[key] = prop
        y += 1
        return proportions
```
