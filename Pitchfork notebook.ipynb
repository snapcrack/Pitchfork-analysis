```python
reviews = pd.read_csv("reviewspandas2.csv", skipinitialspace = True)
```

```python
reviews = reviews.fillna('')
```

```python
reviews['words'] = reviews['content'].str.split()
```

```python
def counting(x):
    counter = {}
    for e in x:
        if e not in counter:
            counter[e] = 0
        if e in counter:
            counter[e] += 1
    return counter
```

```python
reviews['counter'] = reviews['words'].apply(counting)
```

```python
reviews['length'] = reviews['words'].apply(lambda x: len(reviews['words'][x]))
```

```python
##WORD CLEANING

def clean(x):
    cleanwords = []
    for word in x[:]:
        word = update_words(word)
        cleanwords.append(word)
    return cleanwords

def update_words(word):
    clean_word = re.findall(r'(.*)[.]', word)
    clean_word2 = re.findall(r'[A-Z].*', word)
    clean_word3 = re.findall(r'^\W*(\w*)', word)
    clean_word4 = re.findall(r'(\w*)\W*$', word)
    if clean_word:
        cleaned_word = clean_word[0]
        return cleaned_word
    elif clean_word2:
        cleaned_word2 = clean_word2[0].lower()
        return cleaned_word2
    elif clean_word3:
        cleaned_word3 = clean_word3[0].lower()
        return cleaned_word3
    elif clean_word4:
        cleaned_word4 = clean_word4[0].lower()
        return cleaned_word4
    else:
        return word.lower()
```
